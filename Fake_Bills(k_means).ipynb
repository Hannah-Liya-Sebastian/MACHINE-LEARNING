Choosing the Annual Imcome Column & Spending Score Column

```python
x=iris.iloc[:,[2,3]].values
```

```python
print(x)
```

Choosing the number of clusters

WCSS->Within Clusters Sum of Squares

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="random",random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title("The Elbow Point Graph")
plt.xlabel("Numebr of Clusters")
plt.ylabel("WCSS")
plt.show()
```

Optimum Number Of Clusters=3

Training the k-means Clustering Model

```python
kmeans=KMeans(n_clusters=3,init="random",random_state=0)
y=kmeans.fit_predict(x)
print(y)
```

3 Clusters-0,1,2

Visualizing all the clusters

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s=50,c="green",label="Cluster 1")
plt.scatter(x[y==1,0],x[y==1,1],s=50,c="yellow",label="Cluster 2")
plt.scatter(x[y==2,0],x[y==2,1],s=50,c="black",label="Cluster 3")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c="red",label="Centroids")
plt.title("Iris")
plt.xlabel("PetalLengthCm")
plt.ylabel("SepalWidthCm")
plt.show()
```
